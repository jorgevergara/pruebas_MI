# Cálculo de distribución de probabilidad de Laser p(x,y)
## (Se usó kernel Gaussiano (Parzen))

```python
nme = 1000                         # Número de muestras en cada eje para estimar probabilidad
bw  = 5                            # Ancho de banda del kernel gaussiano
x = np.linspace(-10,ns+10,nme)     # Muestreo eje del tiempo
y = np.linspace(-10,255+10,nme)    # Muestreo eje del tiempo
Xtrain = np.vstack([t.ravel(), S.ravel()]).T
X, Y = np.meshgrid(x, y)
xy = np.vstack([X.ravel(), Y.ravel()]).T
kde = kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(Xtrain)
Z = np.exp(kde.score_samples(xy)).reshape(X.shape)
# plot contours of the density
levels = np.linspace(0, Z.max(), 25)
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(1,1,1)
f1 = ax.contourf(X, Y, Z, levels=levels, cmap=plt.cm.Reds)
ax.set_xlabel('Time (Sec.)')
ax.set_ylabel('Amplitude')
cbar = fig.colorbar(f1, cmap=plt.cm.Reds, shrink=0.7)
cbar.ax.set_ylabel('p(x,y)')
```
